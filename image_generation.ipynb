```python
import torch
from diffusers import StableDiffusionPipeline
model_id='runwayml/stable-diffusion-v1-5'
pipe=StableDiffusionPipeline.from_pretrained(model_id,torch_dtype=torch.float16)
pipe=pipe.to('cuda')
```

```python
prompt='a futuristiccity at sunset, cyberpunk style, in great detail'
image=pipe(prompt).images[0]
image
```

```python
image.save('cyberpunk.png')
```

```python
from IPython.display import display

prompts=[
    'a cozy cafe with a warm soft tone on a gloomy day cinematic style',
    'a forest with magical glowing trees in the night in a fantasty world like that of harry potter'
]

for i,p in enumerate(prompts):
  imagee=pipe(p).images[0]
  imagee
  imagee.save(f'image{i+1}.png')
  display(imagee)
```
